```python
import tensorflow as tf
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import numpy as np
import time

# 1. AYARLAR VE VERİ HAZIRLIĞI
batch_size = 256
z_dim = 100
epochs = 50

# MNIST verilerini TensorFlow ile indir ve hazırla
(train_images, train_labels), (_, _) = tf.keras.datasets.mnist.load_data()
train_images = train_images.reshape(train_images.shape[0], 28, 28, 1).astype('float32')
train_images = (train_images - 127.5) / 127.5 # Pikselleri -1 ile 1 arasına sıkıştır

train_dataset = tf.data.Dataset.from_tensor_slices(train_images).shuffle(60000).batch(batch_size)

# 2. ÜRETİCİ AĞ (Generator) - Kalpazan
def uretici_model_olustur():
    model = tf.keras.Sequential([
        layers.Dense(256, use_bias=False, input_shape=(z_dim,)),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.2),

        layers.Dense(512, use_bias=False),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.2),

        layers.Dense(1024, use_bias=False),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.2),

        layers.Dense(28 * 28 * 1, activation='tanh'),
        layers.Reshape((28, 28, 1)) # Sonuç: 28x28 = 784 piksel
    ])
    return model

# 3. AYIRT EDİCİ AĞ (Discriminator) - Dedektif
def ayirt_edici_model_olustur():
    model = tf.keras.Sequential([
        layers.Flatten(input_shape=(28, 28, 1)),
        layers.Dense(512),
        layers.LeakyReLU(alpha=0.2),
        layers.Dense(256),
        layers.LeakyReLU(alpha=0.2),
        layers.Dense(1, activation='sigmoid') # Gerçek(1) veya Sahte(0) tahmini
    ])
    return model

G = uretici_model_olustur()
D = ayirt_edici_model_olustur()

# Hata Hesaplama ve Öğrenme Araçları
cross_entropy = tf.keras.losses.BinaryCrossentropy()
g_optimizer = tf.keras.optimizers.Adam(0.0002, 0.5)
d_optimizer = tf.keras.optimizers.Adam(0.0002, 0.5)

# Kalpazanın Hatası (Dedektifi ne kadar kandırabildiği)
def uretici_hatasi(fake_output):
    return cross_entropy(tf.ones_like(fake_output), fake_output)

# Dedektifin Hatası (Gerçeği ve sahteyi ne kadar doğru ayırdığı)
def ayirt_edici_hatasi(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    return real_loss + fake_loss

# 4. TEK BİR EĞİTİM ADIMI
@tf.function
def egitim_adimi(images):
    noise = tf.random.normal([batch_size, z_dim])

    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        generated_images = G(noise, training=True)

        real_output = D(images, training=True)
        fake_output = D(generated_images, training=True)

        gen_loss = uretici_hatasi(fake_output)
        disc_loss = ayirt_edici_hatasi(real_output, fake_output)

    # Ağırlıkları güncelle (Öğrenme süreci)
    gradients_of_generator = gen_tape.gradient(gen_loss, G.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, D.trainable_variables)

    g_optimizer.apply_gradients(zip(gradients_of_generator, G.trainable_variables))
    d_optimizer.apply_gradients(zip(gradients_of_discriminator, D.trainable_variables))

    return gen_loss, disc_loss

# Sonuçları Kaydetmek İçin Yardımcı Fonksiyon
def ornek_goster(epoch, model):
    noise = tf.random.normal([16, z_dim])
    predictions = model(noise, training=False)
    fig = plt.figure(figsize=(4, 4))
    for i in range(predictions.shape[0]):
        plt.subplot(4, 4, i+1)
        plt.imshow(predictions[i, :, :, 0] * 127.5 + 127.5, cmap='gray')
        plt.axis('off')
    plt.suptitle(f"Epoch {epoch} Sonrası Üretilen Görüntüler")
    plt.show()

# 5. ANA EĞİTİM DÖNGÜSÜ
print("Eğitim başlıyor... (TensorFlow ile MNIST Görüntü Oluşturma)")
for epoch in range(epochs):
    start = time.time()

    for image_batch in train_dataset:
        g_loss, d_loss = egitim_adimi(image_batch)

    print(f"Epoch {epoch+1} Bitti | Süre: {time.time()-start:.1f} sn | Dedektif Hatası: {d_loss:.4f} | Kalpazan Hatası: {g_loss:.4f}")

    # SUNUM İÇİN AYAR: Her 5 turda bir gidişatı gösterir
    if (epoch + 1) % 5 == 0:
        ornek_goster(epoch + 1, G)

print("Eğitim tamamlandı!")
# En son hali tekrar göster
ornek_goster(epochs, G)
```
