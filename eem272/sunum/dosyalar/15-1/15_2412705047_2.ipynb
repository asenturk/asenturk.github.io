```python
# PROJE 1: ETKİLEŞİMLİ ÖĞRENEN ÇEVİRİ SİSTEMİ

# 1. Sistemin Belleğini Tanımlıyoruz
# Bu aşama, Seq2Seq mimarisindeki veri eşleşmesini temsil eder.
ogrenilmis_sistem = {}

def sisteme_ogret(ing_cumle, tr_cumle):
   
    anahtar = ing_cumle.lower().strip()
    deger = tr_cumle.lower().strip()
    ogrenilmis_sistem[anahtar] = deger

# 2. ÖNCELİKLİ ÖĞRETME (Hocaya göstermek için 1-2 metin)
# Kod çalışır çalışmaz bu iki cümleyi hafızaya alacak.
sisteme_ogret("hello world", "merhaba dünya")
sisteme_ogret("I am going to school", "Okula gidiyorum")

# 3. UYGULAMA DÖNGÜSÜ (Input Aşaması)
print("--- YEREL ÇEVİRİ SİSTEMİ BAŞLATILDI ---")
print("İpucu: 'hello world' veya 'I am going to school' yazabilirsiniz.")
print("Çikiş yapmak için 'q' tuşuna basin.\n")

while True:
    kullanici_girdisi = input("İngilizce Metin Girin: ")

    # Çıkış kontrolü
    if kullanici_girdisi.lower() == 'q':
        print("Sistem kapatiliyor...")
        break
    
    # Kelime/Metin kontrolü
    temiz_girdi = kullanici_girdisi.lower().strip()
    
    if temiz_girdi in ogrenilmis_sistem:
        # Öğrenilmişse çeviriyi getir
        print(f"Çeviri: {ogrenilmis_sistem[temiz_girdi]}\n")
    else:
        # Öğrenilmemişse hata ver
        print(f"Hata: '{kullanici_girdisi}' henüz sisteme öğretilmedi! (Veri yok)\n")
```
