```python
# Kütüphaneyi indiriyoruz
%pip install deep-translator
from deep_translator import GoogleTranslator

def cevirmen_uygulamasi():
    print("--- Basit Çeviri Projesi ---")
    print("Çikmak için 'q' tuşuna basiniz.\n")
    
    # Hedef dili seçiyoruz (Örn: İngilizce'den Türkçe'ye)
    translator = GoogleTranslator(source='en', target='tr')
    
    while True:
        metin = input("İngilizce metin girin: ")
        
        if metin.lower() == 'q':
            print("Uygulama kapatiliyor...")
            break
            
        try:
            # Önceden eğitilmiş sistemi kullanarak çeviri yapılıyor
            sonuc = translator.translate(metin)
            print(f"Türkçe Çeviri: {sonuc}\n")
        except Exception as e:
            print(f"Bir hata oluştu: {e}")

# Uygulamayı başlat
cevirmen_uygulamasi()
```
