```
# 1. Önce gerekli kütüphaneleri Colab ortamına kuruyoruz/güncelliyoruz
!pip install --upgrade gensim

import gensim.downloader as api

# 2. Modeli yüklüyoruz (Google Colab'de çok hızlı iner)
print("Model yükleniyor... Lütfen bekleyin.")
model = api.load('word2vec-google-news-300')

# 3. Benzerlik Analizi (Örnek kelimeyi 'engineering' yaptık)
print("\n--- BENZERLİK ANALİZİ ---")
kelime = "engineering"
sonuclar = model.most_similar(kelime, topn=5)

for k, skor in sonuclar:
    print(f"- {k}: %{skor*100:.2f}")

# 4. Analoji Testi (Kral - Erkek + Kadın)
print("\n--- ANALOJİ TESTİ ---")
analoji = model.most_similar(positive=['king', 'woman'], negative=['man'], topn=1)
print(f"Kral - Erkek + Kadın = {analoji[0][0]} (%{analoji[0][1]*100:.2f})")
```
