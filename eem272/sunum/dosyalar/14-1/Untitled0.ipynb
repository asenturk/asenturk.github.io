```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import gensim.downloader as api

# Grafiklerin daha net görünmesi için stil ayarı
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
print("Model yükleniyor... Lütfen bekleyin.")
model = api.load("glove-twitter-25")
print("Model başarıyla yüklendi!")
```

```python
# Farklı kategorilerden kelimeler seçiyoruz
words = [
    "coffee", "tea", "drink", "water", "milk",
    "king", "queen", "man", "woman", "prince",
    "football", "basketball", "soccer", "tennis", "ball",
    "laptop", "computer", "screen", "keyboard", "software",
    "london", "berlin", "paris", "tokyo", "madrid"
]

# Kelimelerin vektörlerini alalım
word_vectors = np.array([model[w] for w in words])

# t-SNE algoritmasını kuralım ve 2 boyuta düşürelim
tsne = TSNE(n_components=2, perplexity=5, random_state=42, init='pca', learning_rate='auto')
word_vectors_2d = tsne.fit_transform(word_vectors)
print("Boyut indirgeme tamamlandı! Grafiğe hazırız.")
```

```python
plt.figure(figsize=(12, 8))

# Her kelimeyi 2D uzayda işaretle
for i, word in enumerate(words):
    x, y = word_vectors_2d[i, 0], word_vectors_2d[i, 1]
    plt.scatter(x, y, edgecolors='k', c='cyan', s=100)
    plt.annotate(word, xy=(x, y), xytext=(8, 2), textcoords='offset points', ha='left', va='bottom', fontsize=12, fontweight='bold')

plt.title("t-SNE ile Kelime Kümeleri (Word Embeddings)", fontsize=16)
plt.xlabel("Boyut 1", fontsize=12)
plt.ylabel("Boyut 2", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
