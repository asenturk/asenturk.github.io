```python
import numpy as np
from numpy.linalg import norm

# Siyam ağından (CNN) çıkmış gibi düşündüğümüz 4 boyutlu temsili yüz vektörleri (Embedding'ler)
# Gerçek projelerde bu vektörler genellikle 128 boyutlu olur.
yuz_A = np.array([4.0, 2.0, 7.0, 1.0]) # Referans yüzümüz (Kişinin sisteme kayıtlı yüzü)
yuz_B = np.array([3.5, 2.5, 6.5, 1.5]) # Kameraya bakan kişinin yüzü (Aynı kişi, hafif farklı ışık)
yuz_C = np.array([-4.0, -2.0, 1.0, -5.0]) # Tamamen farklı birinin yüzü

def cosine_benzerlik(v1, v2):
    # 1. Adım: Vektörlerin nokta çarpımını (dot product) hesapla
    nokta_carpimi = np.dot(v1, v2)
    
    # 2. Adım: Vektörlerin büyüklüklerini (uzunluklarını) hesapla
    buyukluk_v1 = norm(v1)
    buyukluk_v2 = norm(v2)
    
    # 3. Adım: Formülü uygula
    return nokta_carpimi / (buyukluk_v1 * buyukluk_v2)

# Benzerlikleri Test Edelim
benzerlik_AB = cosine_benzerlik(yuz_A, yuz_B)
benzerlik_AC = cosine_benzerlik(yuz_A, yuz_C)

print(f"Aynı kişinin iki fotoğrafı (A ve B) arasındaki benzerlik: {benzerlik_AB:.4f}")
print(f"Farklı kişilerin fotoğrafları (A ve C) arasındaki benzerlik: {benzerlik_AC:.4f}")
```
