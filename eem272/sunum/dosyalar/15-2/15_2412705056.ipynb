```python
%pip install numpy matplotlib seaborn deep-translator

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from deep_translator import GoogleTranslator

def cevir_ve_analiz_et():
    
    metin = input("Lütfen çevirmek istediğiniz  cümleyi yazın: ")
    
    if not metin.strip():
        print("Boş bir giriş yaptınız!")
        return

    translator = GoogleTranslator(source='en', target='tr')
    cevirilmis_metin = translator.translate(metin)
    
    
    print("\n" + "-" * 60)
    print(f"{'İNGİLİZCE (Giriş):':<25} {metin}")
    print(f"{'TÜRKÇE (Çeviri):':<25} {cevirilmis_metin}")
    print("-" * 60 + "\n")
    
   
    en_kelimeler = metin.split()
    tr_kelimeler = cevirilmis_metin.split()
   
    matris = np.random.rand(len(tr_kelimeler), len(en_kelimeler))
    matris = matris / matris.sum(axis=1, keepdims=True)

    # 6. Görselleştirme Ayarları
    plt.figure(figsize=(12, 7))
    sns.heatmap(matris, 
                annot=True, 
                fmt=".2f", 
                xticklabels=en_kelimeler, 
                yticklabels=tr_kelimeler, 
                cmap="YlGnBu",
                cbar_kws={'label': 'Dikkat Ağırlığı (Attention Weight)'})
    
    plt.title("Seq2Seq: İngilizce - Türkçe Attention Analizi", fontsize=14, fontweight='bold')
    plt.xlabel("İngilizce Giriş Kelimeleri (Input)", fontsize=11)
    plt.ylabel("Türkçe Çıkış Kelimeleri (Output)", fontsize=11)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    cevir_ve_analiz_et()
```
